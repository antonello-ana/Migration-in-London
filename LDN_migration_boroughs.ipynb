```python
#importing libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# importing dataset
df = pd.read_csv("/Users/anaantonello/Desktop/Jupter NoteBooks/data/Ldn_polulation_born country.csv")
df.head(5)
```

## Renaming columns and dropping unecessary

```python
df.rename(columns = {'Name':"Borough"}, inplace=True)
df.drop(columns = ['Area Code', 'ALL\n+/- CI'], axis = 1, inplace = True)
df.head(5)
```

## Editing values

### editing strings of a specific cell

```python
df.at[0, 'Geography'] = 'Continent'
df.head(5)
```

### editing numbers

removing commas

replace strings for NA

convert obj to floats


```python
# Remove commas from numbers

for col in df.columns[2:]:
    df[col] = df[col].str.replace(',', '', regex=False)
    
df.head()
```

```python
# Replace text to NA to trnasform numbers in int 

df.replace({"Not applicable": pd.NA, "No contacts": pd.NA, "Confidential": pd.NA}, inplace=True)
df.head()
```

```python
df.info(verbose = True)
```

```python
# Convert columns to numeric type, which will convert non-convertible values to NaN
df_num = df

for col in df.columns[2:]:
    df[col] = pd.to_numeric(df[col], errors='coerce')
    
df_num.head()
```

```python
df_num.info(verbose = True)
```

```python
#set boroughs as index

df_num.set_index('Borough', inplace = True)

df_num = df_num[~df_num.index.isin(['LONDON', 'Inner London', 'Outer London'])]
df_num.sort_index
df_num.head(36)
```

```python
df_num.info(verbose = True)
```

```python
df_num = df_num.dropna(subset = df_num.columns[2:], how= 'all')
df_num.sample(4)
```

```python
# Change columns names

df_num.rename(columns = {'Geography' : 'Region', 'ALL\nEstimate': 'Total'}, inplace = True)


# Remove \Estimate from columns names

df_num.columns = df_num.columns.str.replace('\nEstimate', '')


df_num.head(2)
```

```python
for col in df_num.columns:
    if 'CI' in col:
        del df_num[col]
        
df_num.head(2)
```

```python
df_melted = df_num.reset_index()
df_melted = df_melted.melt(id_vars=['Borough', 'Region', 'Total'], var_name = 'Country of Birth', value_name = 'Estimate')

df_melted = df_melted[['Borough', 'Region', 'Total', 'Country of Birth', 'Estimate']]
df_melted.set_index('Borough', inplace=True)

df_melted['Proportion_estimate'] = df_melted['Estimate']/df_melted['Total']
df_melted.head()
```

```python
df_melted.to_csv("/Users/anaantonello/Desktop/Jupter NoteBooks/data/Ldn_polulation_country_clean.csv")
```

# Top migrants in each borough

#### Sort countries by borough, return the name of the top country

### What are the most prevalent oversea nationality in each borough?

```python
df_melted.head(2)
```

```python
# Calculate the total of immigrants in each borough
total_immigr = df_melted.groupby('Borough')['Estimate'].sum().reset_index()
total_immigr.rename(columns = {'Estimate':'Total_immigr'}, inplace = True)
total_immigr.head()
```

```python
# Add this series back to the df_melted df (merge)
df_melted = pd.merge(df_melted, total_immigr, on = 'Borough')
df_melted.head()
```

```python
# Proportion of immigrants in each borough

df_melted['Prop_immigr'] = df_melted['Total_immigr']/df_melted['Total']
df_melted.head()
```

```python
# Reorder columns

new_cols_order = ['Borough','Region','Total', 'Total_immigr', 'Prop_immigr', 'Country of Birth', 'Estimate', 'Proportion_estimate']
df_melted = df_melted[new_cols_order]
df_melted.head()
```

```python
# Drop NaN from 'Proportion' values
df_melted = df_melted.dropna(subset=['Proportion_estimate'])

# Find the row (index) with the maximum proportion for each borough to filter
max_indices = df_melted.groupby('Borough')['Proportion_estimate'].idxmax()

# Filter the rows with the maximum proportions
max_proportion = df_melted.loc[max_indices]

# Sort the resulting DataFrame by 'Proportion' in descending order
max_proportion = max_proportion.sort_values(by='Proportion_estimate', ascending=False).reset_index(drop = True)

# Display the DataFrame with the maximum proportions
max_proportion.head(5)
```

```python
# Drop NaN from 'Proportion' values
df_melted = df_melted.dropna(subset=['Proportion_estimate'])

# Find the row (index) with the maximum proportion for each borough to filter
max_indices = df_melted.groupby('Borough')['Proportion_estimate'].idxmax()

# Filter the rows with the maximum proportions
max_proportion = df_melted.loc[max_indices]


# Sort the resulting DataFrame by 'Proportion' in descending order
max_proportion = max_proportion.sort_values(by='Proportion_estimate', ascending=False).reset_index(drop = True)

# Display the DataFrame with the maximum proportions
max_proportion.head(5)
```

```python
colors = {'Outer London' : "#9ecae1", "Inner London" : "#3182bd"}

plt.figure(figsize = (6,10))

bar_max = plt.barh(max_proportion['Borough'], max_proportion['Proportion_estimate'], color = max_proportion['Region'].map(colors))

# Titles
plt.ylabel('Borough of London')
plt.title('Top immigrant nationality in each borough of London')

# Add the value to the bar
for bar, country, proportion in zip(bar_max, max_proportion['Country of Birth'], max_proportion['Proportion_estimate']):
    plt.text(bar.get_width(), bar.get_y() + bar.get_height()/2, f'{country} - {proportion: .1%}',
             va = 'center', ha = 'left', fontsize = 8, color = 'black')

plt.gca().invert_yaxis()

plt.show()
```

```python
# Calculate the remaining proportion to complete 100% in 2 new columns (['Immigrant_remaining'] and ['total_remaining'])

max_proportion['Immigrant_remaining'] = max_proportion['Prop_immigr'] - max_proportion['Proportion_estimate']
max_proportion['total_remaining'] = 1 - max_proportion['Prop_immigr']
max_proportion.head()
```

```python
colors = {'Outer London': "#9ecae1", "Inner London": "#3182bd"}


plt.figure(figsize=(8, 10))

# Plot Proportion bars (blue)
bar_max = plt.barh(max_proportion['Borough'], max_proportion['Proportion_estimate'], color = max_proportion['Region'].map(colors))

# Plot remaining bars (greys)
# Dark grey for total of immigrants
plt.barh(max_proportion['Borough'], max_proportion['Immigrant_remaining'], left = max_proportion['Proportion_estimate'], color = '#d9d9d9')

# Light Grey for total of population
plt.barh(max_proportion['Borough'], max_proportion['total_remaining'], left = max_proportion['Prop_immigr'], color = '#f0f0f0')


# Titles and labels

plt.ylabel('Borough of London')
plt.title('Top immigrant nationality in each borough of London')

# Add the value to the bar
for bar, country, proportion in zip(bar_max, max_proportion['Country of Birth'], max_proportion['Proportion_estimate']):
    plt.text(bar.get_width(), bar.get_y() + bar.get_height() / 2, f'{country}: {proportion:.1%}',
             va='center', ha='left', fontsize=8, color='black')

# Add a legend for inner and outer london regions
plt.legend(handles=[plt.Rectangle((0, 0), 1, 1, color=colors['Outer London']),
                    plt.Rectangle((0, 0), 1, 1, color=colors['Inner London'])],
           labels=['Outer London', 'Inner London'])

# To remove the edges
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['bottom'].set_visible(False)
plt.gca().spines['left'].set_visible(False)

plt.gca().invert_yaxis()
plt.xticks([])

plt.tight_layout()
plt.show()
```

# How is the proportion distribution of immigrants? 

Back to use df_melted df because this has all the country, not only the top immigrants ones (max_proportion)

```python
df_melted['median_proportion'] = df_melted.groupby('Borough')['Proportion_estimate'].transform('median')
df_melted.sort_values(by = ['median_proportion', 'Borough'], ascending = False, inplace = True)
df_melted.head(2)
```

```python
borough_order = df_melted['Borough'].unique()
borough_order 
```

```python
import seaborn as sns
from matplotlib.lines import Line2D  # This is a thing for the legend


plt.figure(figsize= (8,20))

p = sns.swarmplot(x = 'Proportion_estimate', y = 'Borough', data = df_melted, 
                  size = 4,
                 zorder = 10)


# plot the median line
sns.boxplot(showmeans=False,
            meanline=False,
            medianprops={'visible': True, 'color': 'grey', 'ls': '-', 'lw': 2},
            whiskerprops={'visible': False},
            zorder=1,
            x="median_proportion",
            y="Borough",
            data=df_melted,
            showfliers=True,
            showbox=False,
            showcaps=False,
            ax = p)

plt.title("Proportion of overseas nationalities in each borough of London")
plt.xlabel('Proportion of Immigrants')
plt.ylabel('')

# Adding a custom legend for the median line
median_line = Line2D([0], [0], color='grey', linewidth=2, label='Median')
plt.legend(handles=[median_line], loc='upper right')


plt.show()
```

## Interactive chart

```python
from bokeh.plotting import figure, output_notebook, show

from bokeh.transform import jitter

from bokeh.models import HoverTool, ColumnDataSource


# Ensure Bokeh plots will display in the notebook
output_notebook()
```

```python
#df.info(verbose = True)
```

```python
median_values = df_melted.groupby('Borough')['Proportion_estimate'].median().reset_index()


# Create a Bokeh figure
p = figure(width=600, height=1500, y_range=borough_order.tolist(), title="Proportion by Borough")


# Prepare data source for Bokeh --> Bokeh needs lists, ColumnDataSource treat df as lists of each column
source = ColumnDataSource(df_melted)


# Plot the swarmplot using circles
p.circle(x='Proportion_estimate', y=jitter('Borough', width=0.2, range=p.y_range), source=source, size=6, color='navy', alpha=0.3)



# Toolpit when hover the mouse over each circle
hover = HoverTool(tooltips=[
    ("Country", "@{Country of Birth}"),
    ("Borough", "@Borough"),
    ("Proportion", "@Proportion_estimate{0.2f}%")])

p.add_tools(hover)


# Labels
p.xaxis.axis_label = 'Proportion_estimate'
p.yaxis.axis_label = 'Borough'
p.title.align = "center"

show(p)

```
